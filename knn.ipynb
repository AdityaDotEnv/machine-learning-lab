# K-Nearest Neighbors Algorithm

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier

print("\n--- K- Nearest Neighbors Classifier ---")
X = np.random.rand(100, 1) # Generate 100 random numbers in the range 0 to 1
y = np.array([1 if x <= 0.5 else 2 for x in X[:50]]) # Split random numbers into two classes

X_train, y_train = X[:50], y # 50 generated numbers for training 
X_test = X[50:] # Remaining 50 numbers for testing
y_test = np.array([1 if x <= 0.5 else 2 for x in X_test])
acc = []
kv = [1, 2, 3, 4, 5, 20, 30] # K values to be tested for as specified in the problem statement

for k in kv:
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    acc.append(accuracy)
    print(f"k = {k}")
    print(f"Accuracy: {accuracy}")

plt.plot(kv, acc, marker='o', linestyle='-', color='c')
plt.show()
```
