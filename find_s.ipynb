# Find-S Algorithm Implementation

```python
import csv

filename = "training_data.csv"

# Function to implement Find-S Algorithm
def find_s_algorithm(file):
    
    # Read the CSV file
    with open(file,'r') as f:
        reader = csv.reader(f)
        data = list(reader)

    # Extract attributes names (excluding target columns)
    attributes = data[0][:-1]

    # Extract training examples
    examples = data[1:]

    # Initialize hypothesis with most specific values
    hypothesis = ['0'] * len(attributes)

    # Process each training example
    for example in examples:

        # Consider only possible examples
        if example[-1].strip().lower() == 'yes':
            for i in range(len(attributes)):
                if hypothesis[i] == '0':
                    hypothesis[i] = example[i]
                elif hypothesis[i] != example[i]:
                    hypothesis[i] = '?'

    return hypothesis

# Run the algorithm using the file names
final_hypothesis = find_s_algorithm(filename)

# Display the result
print("Training file used:", filename)

print("Final Hypothesis:", final_hypothesis)

```
